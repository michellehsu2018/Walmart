```python
import pandas as pd
from collections import Counter
```

```python
train = pd.read_csv("train.csv",dtype = {'Upc': str, 'FinelineNumber':"category", 'TripType':"category" , 'Weekday':"category"})
test = pd.read_csv("test.csv",dtype = {'Upc': str, 'FinelineNumber':"category", 'TripType':"category" , 'Weekday':"category"})
```

```python
all_data = pd.concat([train,test])
all_data.reset_index(inplace=True,drop=True)
```

```python
def checksum(x):
    try:
        odd = map(int, ','.join(x[-1::-2]).split(',')) # ''.join(sequence) # fast way to concatenate a sequence of string
        even = map(int, ','.join(x[-2::-2]).split(','))
        sum_odd3 = sum(odd) * 3
        total = sum_odd3 + sum(even)
        rem = total % 10
        if rem == 0:
            return rem
        return 10 - rem
    except:
        return -9999
```

```python
## add the checksum to all the UPCs
add_checksum = []
for i in range(len(all_data)):
    upc_str = str(all_data['Upc'][i])
    if(len(upc_str)<12):
        check_sum = checksum(upc_str)
        if check_sum == -9999:
            add_checksum.append(str(check_sum))
        else:
            new_upc = upc_str + str(check_sum)
            add_checksum.append(new_upc) 
    else:
        add_checksum.append(upc_str)
```

```python
add_checksum
```

```python
def full_npc(x):
    if (len(x) < 12) and (x != '-9999'):
        missing = 12 - len(x)
        zeros = ['0'] * missing
        xx = zeros + ','.join(x).split(',')
        xx = ''.join(xx)
        return xx
    else:
        return x
```

```python
test = []
for i in range(len(add_checksum)):
    full = full_npc(add_checksum[i])
    test.append(full)
```
